## Auto Theft occurrences by reported date.
This dataset includes all auto theft occurrences by reported date and related offences since 2014.This data is provided at the offence and/or vehicle level, therefore one occurrence number may have several rows of data associated to the various MCIs used to categorize the occurrence.The downloadable datasets display the REPORT_DATE and OCC_DATE fields in UTC timezone.This data does not include occurrences that have been deemed unfounded. The definition of unfounded according to Statistics Canada is: “It has been determined through police investigation that the offence reported did not occur, nor was it attempted” (Statistics Canada, 2020).
The dataset is intended to provide communities with information regarding public safety and awareness. The data supplied to the Toronto Police Service by the reporting parties is preliminary and may not have been fully verified at the time of publishing the dataset.The location of crime occurrences have been deliberately offset to the nearest road intersection node to protect the privacy of parties involved in the occurrence. All location data must be considered as an approximate location of the occurrence and users are advised not to interpret any of these locations as related to a specific address or individual.

```python
from scipy.stats import chi2_contingency
def chi_square(c1,c2):
    chi_2, p_val, dof, exp_val = chi2_contingency(pd.crosstab(Theft[c1],Theft[c2],margins = False))# make sure margins = False
    print("Actual values:\n")
    print(pd.crosstab(Theft[c1],Theft[c2],margins = False))
    print("="*100)
    print("Expected values: \n")
    print(exp_val)
    print("="*100)
    #print('\nChi-square is : %f'%chi_2, '\n\np_value is : %f'%p_val, '\n\ndegree of freedom is : %i'%dof)
    print(f'\nChi-square is : {chi_2}', f'\n\np_value is : {p_val}', f'\n\ndegree of freedom is :{dof}')
    print("="*100)
    if exp_val.min()<5 or np.sum(exp_val)<50 : print("\n********************************************\nWarning :We have some small expected \
values or the total number of observations is not enough to get any conclusion\n******************************************** ")
        #Note for sum of all values in dataframe use df.values.sum() and for numpy.ndarray use np.sum(arr)
    elif p_val < 0.05:# consider significan level is 5%
        print(F"\nThere is statistiacally significant correlation between {c1} and {c2} at 0.05 significant level")
    else:
        print(F"\nThere is no correlation between the two variables( we don't have enough evidence to conclude there is a a statistically significant relationship between {c1} and {c2}")        
```

✅ Assumptions of the Chi-Square Test
The Chi-Square Test (commonly used in independence or goodness-of-fit tests) has several important assumptions. These must be met for the results to be valid and interpretable.

📌 1. Variables Should Be Categorical
Both variables involved should be nominal or ordinal (not continuous).

You can’t use chi-square directly on continuous variables unless they are binned into categories.

📌 2. Independence of Observations
Each observation (row in your data) should be independent of all others.

No repeated measures or paired samples.

If your data includes repeated measures or clustering, use other tests like McNemar's or Cochran’s Q.

📌 3. Adequate Expected Frequencies
Expected frequency in each cell of the contingency table should be at least 5.

For 2x2 tables: All expected counts should be ≥ 5.

For larger tables: No more than 20% of cells should have expected counts < 5.

If this is violated, use Fisher’s Exact Test (for small samples or 2x2 tables).

📌 4. Random Sampling
The data should be drawn from a population using random sampling methods.

Non-random sampling can introduce bias and limit generalizability.

📌 5. Sample Size
The Chi-square test is a large sample test.

Works best when sample size is sufficient — very small samples can lead to unreliable results.

## Assumptions of ANOVA

### 1. Independence: The groups are mutually exclusive so that means this assumption is met.

### 2. Normality: The residuals (errors) from the ANOVA model should be approximately normally distributed. 
Lets check but even this assumptions fails then also, ANOVA is fairly robust to violations of normality if sample sizes are equal and large (Central Limit Theorem applies, which is met)

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fit ANOVA model using OLS
model = sm.formula.ols('REPORT_HOUR ~ C(PREMISES_TYPE)', data=Theft).fit()

# Q-Q plot of residuals
sm.qqplot(model.resid, line='s')
plt.title('Q-Q Plot of ANOVA Residuals')
plt.show()
```

### 3.Homogeneity of Variances (Homoscedasticity):The variances within each group should be approximately equal.

```python
from scipy.stats import levene
# Split data into groups based on Air_Quality_Level
grouped = Theft.groupby('PREMISES_TYPE')['REPORT_HOUR']

# Extract values from each group
groups = [group for name, group in grouped]

# Run Levene's test
stat, p = levene(*groups)

print(f"Levene’s test statistic: {stat:.4f}")
print(f"p-value: {p:.4f}")
```

```python
import pingouin as pg

# Welch's ANOVA (robust to unequal variances)
welch_result = pg.welch_anova(dv='REPORT_HOUR', between='PREMISES_TYPE', data=Theft)
print(welch_result)
```

F: The F-statistic — 1525.17, indicating the ratio of variance between the group means to the variance within the groups.

p-unc: The uncorrected p-value — 0.00, strongly suggesting statistically significant differences between at least some group means.

np2: Partial eta-squared — 0.10, which is a measure of effect size. According to common benchmarks:

Small effect ≈ 0.01

Medium effect ≈ 0.06

Large effect ≈ 0.14

So, 0.10 indicates a medium-to-large effect size — the type of premises has a notable influence on the reported hour of theft.

Conclusion:
The Welch’s ANOVA result shows that there is a statistically significant difference in REPORT_HOUR among the different PREMISES_TYPE categories, and the effect size is moderate to large.

### Hence, despite the violation of equal variances (from Levene's test), Welch's ANOVA confirms significant differences between PREMISES_TYPE in terms of REPORT_HOUR.

```python
def adjusted_r2(r2, n, p):
    return 1 - (1 - r2) * ((n - 1) / (n - p - 1))
```

```python
def get_p_values(X, y):
    X_with_const = sm.add_constant(X)
    model = sm.OLS(y, X_with_const).fit()
    return model.pvalues
```

```python
results = []

def store_results(model_name, r2, adj_r2, rmse):
    results.append({
        'Model': model_name,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse
    })
```

```python
def evaluate_model(model, model_name, X_test, y_test, X_train):
    predictions = model.predict(X_test)
    r2 = r2_score(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    adj_r2 = adjusted_r2(r2, X_test.shape[0], X_test.shape[1])
    store_results(model_name, r2, adj_r2, rmse)
    print(f"{model_name} -> R2: {r2:.4f}, Adjusted R2: {adj_r2:.4f}, RMSE: {rmse:.4f}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def calculate_metrics(model, X_train, X_test, y_train, y_test):
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    adj_r2 = 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return r2, adj_r2, rmse
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

def evaluate_linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)

    r2, adj_r2, rmse = calculate_metrics(model, X_train, X_test, y_train, y_test)

    # Statsmodels for p-values
    X_train_sm = sm.add_constant(X_train)
    ols_model = sm.OLS(y_train, X_train_sm).fit()

    return {
        'model': model,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse,
        'p-values': ols_model.pvalues,
        'summary': ols_model.summary()
    }
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

def evaluate_knn(X_train, X_test, y_train, y_test, cv_folds=5):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('knn', KNeighborsRegressor())
    ])
    param_grid = {'knn__n_neighbors': [3, 5, 7]}

    grid = GridSearchCV(pipeline, param_grid, cv=cv_folds, scoring='r2', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_

    r2, adj_r2, rmse = calculate_metrics(best_model, X_train, X_test, y_train, y_test)

    return {
        'model': best_model,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse
    }
```

```python
from sklearn.svm import SVR

def evaluate_svr(X_train, X_test, y_train, y_test, cv_folds=5):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('svr', SVR())
    ])
    param_grid = {
        'svr__C': [1, 10],
        'svr__kernel': ['rbf', 'linear']
    }

    grid = GridSearchCV(pipeline, param_grid, cv=cv_folds, scoring='r2', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_

    r2, adj_r2, rmse = calculate_metrics(best_model, X_train, X_test, y_train, y_test)

    return {
        'model': best_model,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse
    }
```

```python
from sklearn.tree import DecisionTreeRegressor

def evaluate_decision_tree(X_train, X_test, y_train, y_test, cv_folds=5):
    model = DecisionTreeRegressor()
    param_grid = {'max_depth': [5, 10, None]}

    grid = GridSearchCV(model, param_grid, cv=cv_folds, scoring='r2', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_

    r2, adj_r2, rmse = calculate_metrics(best_model, X_train, X_test, y_train, y_test)

    return {
        'model': best_model,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse
    }
```

## Compare all the models 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

def fit_and_evaluate_model(model_name, model, param_grid, X_train, X_test, y_train, y_test, cv_folds=5, scale=False):
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import StandardScaler

    # Add scaling if needed
    if scale:
        model = Pipeline([
            ('scaler', StandardScaler()),
            (model_name.lower().replace(" ", "_"), model)
        ])
        # Adjust param grid keys
        param_grid = {
            f"{model_name.lower().replace(' ', '_')}__{k}": v for k, v in param_grid.items()
        }

    # Fit model with grid search if parameters provided
    if param_grid:
        grid = GridSearchCV(model, param_grid, cv=cv_folds, scoring='r2', n_jobs=-1)
        grid.fit(X_train, y_train)
        best_model = grid.best_estimator_
    else:
        model.fit(X_train, y_train)
        best_model = model

    # Evaluate
    y_pred = best_model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    adj_r2 = 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    return {
        'Model': model_name,
        'R2': r2,
        'Adjusted R2': adj_r2,
        'RMSE': rmse,
        'Fitted Model': best_model
    }
```

```python
all_results = []

# Linear Regression (no scaling or hyperparameters)
from sklearn.linear_model import LinearRegression
result = fit_and_evaluate_model(
    model_name='Linear Regression',
    model=LinearRegression(),
    param_grid={},
    X_train=X_train, X_test=X_test,
    y_train=y_train, y_test=y_test
)
all_results.append(result)
```

## plot the regression model results- Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_model_comparison(results_list):
    # Extract relevant metrics from results
    summary_df = pd.DataFrame([{
        'Model': res['Model'],
        'R2': res['R2'],
        'Adjusted R2': res['Adjusted R2'],
        'RMSE': res['RMSE']
    } for res in results_list])
    
    # Melt for easier plotting
    melted = summary_df.melt(id_vars='Model', var_name='Metric', value_name='Score')
    
    # Plot
    plt.figure(figsize=(12, 6))
    sns.barplot(data=melted, x='Model', y='Score', hue='Metric')
    plt.title('Model Performance Comparison')
    plt.xticks(rotation=45)
    plt.xlabel('')
    plt.ylabel('Score')
    plt.legend(title='Metric')
    plt.tight_layout()
    plt.show()
```

```python
plot_model_comparison(all_results)
```
